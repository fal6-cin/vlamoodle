```python
import pandas as pd

all_dfs = []

files_to_join = [
    'datasets/logstore_standard_log_courseid_993_26_03.csv',
    'datasets/logstore_standard_log_courseid_993_26_04.csv',
    'datasets/logstore_standard_log_courseid_993_26_05.csv',
    'datasets/logstore_standard_log_courseid_993_26_06.csv'
]

# Carregar cada arquivo e adicionar à lista 'all_dfs'
for file_name in files_to_join:
    df_temp = pd.read_csv(file_name, sep=';')
    all_dfs.append(df_temp)

# Concatenar todos os dataframes em um único 'logstore'
logstore = pd.concat(all_dfs, ignore_index=True)

logstore.head()
```
